## To find the number of free apps and paid apps


### Creating a new column 'payment_type' and categorizing all the applications into 'free' or 'paid'

```python
# creating a custom function to classify an app as paid or free
def paid_or_free(price):
    """
    Checks if the app paid or free
    
    Params:
    -------
    price: int or float
    
    Returns:
    --------
    free if price is 0 else paid
    """
    
    assert type(price) in [int, float], "You need to enter a number"
    assert price >= 0, "Enter a non negative number"
    
    return 'free' if price == 0 else 'paid'
```

```python
# applying the above created function to the 'price' column of apple DataFrame using map function and storing the returned
# values in a new column called 'payment_type'
apple['payment_type'] = apple['price'].map(paid_or_free)
```

```python
apple.head()
```

```python
# checking the percentage of free and paid apps using the newly created column
apple['payment_type'].value_counts(normalize=True)
```

## Does the price of the app increase with the increase in its size?

#### Creating a new column "size_mb" converting bytes to megabytes.

```python
# Converting bytes to megabytes
def size(x):
    mb = x/(1024**2)
    return mb
    
top_five_genre['size_mb'] = apple['size_bytes'].map(size)

top_five_genre[['size_bytes', 'size_mb']]
```

```python
# creating a scatter plot using data for top 5 genre to see if there is any relation between price and size of apps

plt.figure(figsize = (15,10))
paid_apps_desc = top_five_genre[top_five_genre['price'] <50]

sns.lmplot(data = paid_apps_desc, x = 'size_mb', y = 'price', col = 'prime_genre',col_wrap= 3,aspect= 1.5
           ,scatter = True,fit_reg = False, hue  = 'prime_genre', legend_out = True)
plt.show()
```

* The above plots for top 5 genre suggest that there exists no relation between app size and price. That is the price doesn't increase with an increasein size.

## Can we create a recommendation column based on "user_rating" and "rating_count_tot"?

```python
def recommendation(total_reviews,rating):
# creating an empty list to store the recommendation
    recommendation = []
# Iterating through all the rating and review count values, using the if else statement and categorizing the app
    for review_count,avg_rating in zip(apple['rating_count_tot'],apple['user_rating']):
        if review_count > total_reviews and avg_rating > rating:
            recommendation.append('yes')
        else:
            recommendation.append('no')
# Creating a new recommendation column
    apple['recommendation'] = recommendation
```

```python
# we set the minimum number of reviews required to reccommend an app to 20,000 and the minimum user rating required to 3.5
recommendation(20000,3.5)
```
